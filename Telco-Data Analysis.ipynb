## 4.Testing Null Hypothesis

```python
from scipy.stats import ttest_ind
import numpy as np

df_telco[['MonthlyCharges','Churn']].groupby('Churn').agg(['mean', 'std'])

```

```python
MT_Churn_YES=df_telco[df_telco['Churn']=="Yes"]
MT_Churn_YES=MT_Churn_YES['MonthlyCharges'].to_numpy()

MT_Churn_NO=df_telco[df_telco['Churn']=="No"]
MT_Churn_NO=MT_Churn_NO['MonthlyCharges'].to_numpy()
```

```python
ttest,pval = ttest_ind(MT_Churn_YES,MT_Churn_NO)

print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```
